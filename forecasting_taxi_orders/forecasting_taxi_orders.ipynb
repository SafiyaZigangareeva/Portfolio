Так как ВР имел восходящий тренд, был рассчитан критерий Дики-Фуллера, который составил 0.03, что при уровне значимости 0.05 позволяет считать наш временной ряд стационарным.

### Создание признаков

```python
#Функции для создания периодических признаков с помощью тригонометрических функций
def sin_transformer(period):
    return FunctionTransformer(lambda x: np.sin(x / period * 2 * np.pi))

def cos_transformer(period):
    return FunctionTransformer(lambda x: np.cos(x / period * 2 * np.pi))
```

```python
def make_features(df, max_lag, rolling_mean_size):
    
    """Функция для создания признаков (календарные, отстающие значения, скользящее среднее). Принимает на вход данные, 
    количество шагов для отстающих значений, размер окна для скользящего среднего. Возвращает датасет с новыми признаками."""
    
    df['hour'] = df.index.hour#.astype(str)
    df['dayofweek'] = df.index.dayofweek#.astype(str)
    df['hour_sin'] = sin_transformer(24).fit_transform(df)['hour']
    df['hour_cos'] = cos_transformer(24).fit_transform(df)['hour']
    df['weekday_sin'] = sin_transformer(7).fit_transform(df)['dayofweek']
    df['weekday_cos'] = cos_transformer(7).fit_transform(df)['dayofweek']
    df = df.drop(['hour', 'dayofweek'], axis=1)
    
    for lag in range(1, max_lag + 1):
        df[f'lag_{lag}'] = df['num_orders'].shift(lag)

    df['rolling_mean'] = df['num_orders'].shift().rolling(rolling_mean_size).mean()
    return df
```

```python
df = make_features(df, 168, 168)
df.head()
```

## Обучение

```python
def model_selection(grid, grid_dict):
    
    """Принимает на вход сетку гиперпараметров для каждого класса моделей и соответствующий словарь с названиями моделей. 
    Возвращает параметры лучшей модели в каждом классе."""
    
    result = []
    
    for i in range(len(grid_dict)):
        grid[i].fit(train[features], train[target])
        result.append([grid_dict[i], abs(grid[i].best_score_), grid[i].best_params_])
        
    return pd.DataFrame(result, columns = ['model', 'RMSE', 'params'])
```

```python
features = df.drop('num_orders', axis=1).columns
target = 'num_orders'
```

```python
train, test = train_test_split(df, shuffle = False, test_size = TEST_SIZE, random_state=RANDOM_STATE)
train = train.dropna() 
```

```python
# 1. Проверим модель на адекватность константным прогнозом
dummy_regr = DummyRegressor(strategy="mean")
dummy_regr.fit(train[features], train['num_orders'])

mean_squared_error(test['num_orders'], dummy_regr.predict(test[features]), squared=False)
```

```python
# 2. Проверим модель на адекватность - новое значение прогнозируется предыдущим значением ряда
pred_previous = test['num_orders'].shift()
pred_previous[0] = train['num_orders'].iloc[-1]

mean_squared_error(test['num_orders'], pred_previous, squared=False)
```

```python
#Подготавливаем и проверяем разбиение данных для кросс-валидации
time_split = TimeSeriesSplit(n_splits=5)
[(el[0].shape, el[1].shape) for el in time_split.split(train[features])]
```

```python
#Линейная регрессия
params = {'fit_intercept':[True, False]}

lr = GridSearchCV(LinearRegression(), params, scoring='neg_root_mean_squared_error', cv=time_split, error_score='raise')

params = {'n_estimators':[200],
         'max_depth': [i for i in range(12, 14, 2)],
         'min_samples_split': [i for i in range(2, 4)],
         'min_samples_leaf': [i for i in range(2, 4)],
         'n_jobs': [-1]}

#Случайный леc
rf = GridSearchCV(RandomForestRegressor(random_state=RANDOM_STATE), params, scoring='neg_root_mean_squared_error', 
                  cv=time_split, error_score='raise')

#Градиентный бустинг
params = {'learning_rate': [0.11, 0.12, 0.13],
          'max_depth' : [3, 4, 5],
          'reg_alpha':[0.011, 0.012, 0.013, 0.014],
          'num_leaves':[29, 30, 31],
          'min_child_samples':[16, 17, 18],
          'metric':['rmse'],
          'n_jobs': [-1]}

gbr = GridSearchCV(lgb.LGBMRegressor(random_state=RANDOM_STATE), params, scoring='neg_root_mean_squared_error', 
                   cv=time_split, error_score='raise')


grid = [lr, rf, gbr]
grid_dict = {0: 'LinearRegression', 1:'RandomForestRegressor', 2:'LGBMRegressor'}
```

```python
result = model_selection(grid, grid_dict)
result
```

Лучшая модель по результатам кросс-валидации - LinearRegression с результатом RMSE 22.45. Проверим ее на тестовых данных.

## Тестирование

```python
def f_importances(coef, names, top=-1):
    
    """Функция для определения наиболее важных признаков с точки зрения модели"""
    
    imp = coef
    imp, names = zip(*sorted(list(zip(imp, names))))

    if top == -1:
        top = len(names)
    plt.figure(figsize=(12,5))
    plt.barh(range(top), imp[::-1][0:top], align='center')
    plt.yticks(range(top), names[::-1][0:top])
    plt.show()
```

```python
#Метрика на test
print(f'RMSE на test {mean_squared_error(test[target], lr.predict(test[features]), squared=False)}')
```

```python
f_importances(abs(lr.best_estimator_.coef_), features, top=10)
```
