## Описание проекта  и цель

Чтобы оптимизировать производственные расходы, металлургический комбинат ООО «Так закаляем сталь» решил уменьшить потребление электроэнергии на этапе обработки стали. Предстоит построить модель, которая предскажет температуру стали.

## Описание этапа обработки

Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша. 

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. *slab*, «плита»).

Это ошибочные данные, их нужно удалить. Так же нужно изменить тип столбцов 'Начало нагрева дугой' и 'Конец нагрева дугой' на datetime.

```python
# Удалим из датафрейма строки с партией 2116 
df['data_arc'] = df['data_arc'].loc[df['data_arc']['key'] != 2116]
```

```python
# Изменим тип данных в столбцах на datetime
df['data_arc']['Начало нагрева дугой'] = pd.to_datetime(df['data_arc']['Начало нагрева дугой'], format='%Y-%m-%d %H:%M:%S')
df['data_arc']['Конец нагрева дугой'] = pd.to_datetime(df['data_arc']['Конец нагрева дугой'], format='%Y-%m-%d %H:%M:%S')
```

```python
df['data_arc'].dtypes
```

```python
# Определим время, в течении которого сплав нагревался
df['data_arc']['Время нагрева'] = (df['data_arc']['Конец нагрева дугой'] - df['data_arc']['Начало нагрева дугой']).dt.seconds
```

```python
# Сгруппируем данные для дальнейшего использования
result = df['data_arc'].groupby('key').agg({'Активная мощность' : 'sum', 
                                    'Реактивная мощность' : 'sum', 
                                    'Время нагрева' : 'sum'})
result['Количество нагревов'] = df['data_arc'].groupby('key')['Активная мощность'].count()
result.columns = ['Суммарная активная мощность', 'Суммарная реактивная мощность', 
                    'Суммарное время нагрева', 'Количество нагревов']
result.head()
```

```python
# Добавим в датасет признаки, которые связывают активную и реактивную мощности
result['Полная мощность'] = (result['Суммарная активная мощность']**2 + result['Суммарная реактивная мощность']**2)**0.5
result['Коэффициент мощности'] = result['Суммарная активная мощность']/result['Полная мощность']
result = result.drop(['Суммарная активная мощность', 'Суммарная реактивная мощность'], axis = 1)
result.head()
```

```python
plt.figure(figsize=(4, 4))
result['Коэффициент мощности'].hist()
plt.title("Распределение коэффициента мощности")
plt.xlabel('Коэффициент мощности')
plt.show() 
```

Для расчета полной мощности я воспользовалась формулой, которая выражает ее через активную мощность (P) и реактивную мощность (Q):

$$S = sqrt(P^2 + Q^2)$$

Коэффициент мощности является отношением активной мощности (P) к полной мощности (S):

$$ λ = P/S$$

Полная мощность является мерой всей мощности, потребляемой электрической системой, включая ту, которая не выполняет полезную работу. В электрической системе важно иметь высокий коэффициент мощности, чтобы минимизировать потери мощности, улучшить эффективность системы и сократить расходы на электроэнергию.
На распределении видно, что для большинства наблюдений коэффициент мощности высокий - более 0.8, но так же есть наблюдения с низким коэффициентом 0.6-0.7

## 3. Подбор и обучение модели

```python
def f_importances(coef, top=-1):
    
    """Функция для определения наиболее важных признаков с точки зрения модели"""
    
    # Определим индекс признака с максимальным коэффициентом для каждой компоненты
    model = PCA(n_components=33).fit(X_train)
    most_important = [np.abs(model.components_[i]).argmax() for i in range(model.components_.shape[0])]
    
    # Получим названия признаков с максимальным вкладом для каждой компоненты
    most_important_names = [X_train.columns[most_important[i]] for i in range(model.components_.shape[0])]
    
    li = list(zip(most_important_names, coef, abs(coef)))
    names, coef, abs_coef = zip(*sorted(li, key=lambda x: x[2]))

    if top == -1:
        top = len(names)
    plt.figure(figsize=(10,5))
    plt.barh(range(top), coef[::-1][0:top], align='center')
    plt.yticks(range(top), names[::-1][0:top])
    plt.show()
```

Оценена важность признаков с помощью построения графика feature importances модели RandomForestClassifier, обученной на выборке train. Данная модель наиболее высоко оценила признаки: Коэффициент мощности, Wire 1, Начальная температура, Газ1, Общее рвемя нагрева.

Принято решение использовать для обучения моделей все признаки. Использовать модели, которые устойчивы к мультиколлинеарности, такие как Ridge или Lasso регрессия. Для уменьшения признакового пространства будем использовать PCA - метод главных компонент.
