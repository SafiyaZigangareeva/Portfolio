# Матчинг товаров

## Описание проекта  и цель

- разработать алгоритм, который для всех товаров из validation.csv предложит несколько вариантов наиболее похожих товаров из base;
- оценить качество алгоритма по метрике accuracy@5.

## Описание данных 

- *base.csv* - анонимизированный набор товаров. Каждый товар представлен как уникальный id (0-base, 1-base, 2-base) и вектор признаков размерностью 72.
- *target.csv -* обучающий датасет. Каждая строчка - один товар, для которого известен уникальный id (0-query, 1-query, …) , вектор признаков И id товара из *base.csv*, который максимально похож на него (по мнению экспертов).
- *validation.csv* - датасет с товарами (уникальный id и вектор признаков), для которых надо найти наиболее близкие товары из *base.csv*
- *validation_answer.csv* - правильные ответы к предыдущему файлу.

## План работы

1. Открыть файлы и ознакомится с ними
2. 1 этап - faiss
 - Обучаем различные индексы и рассчитываем accuracy@k
 - Генерируем предсказания Faiss как кандидатов для второго этапа
3. 2 этап - Catboost
 - подбор гиперпараметров, обучение модели
4. Проверка на test моделей обоих этапов:
 - отдельно модели 1го уровня: faiss
 - отдельно двухуровненой модели: faiss + Catboost
5. Выводы

### 1️⃣ 1 этап: Faiss

#### Функции

```python
def calc_acc(idx, targets, base_index):
    """Функция для расчета метрики accuracy@5.
    Принмает на вход индексы товаров-соседей, правильные ответы - максимально соответствующие запросу товары, 
    словарь соотвествия товаров из base и их индексов"""
    acc = 0
    for target, el in zip(targets.values.tolist(), idx.tolist()):
        acc += int(target in [base_index[r] for r in el])
    return 100 * acc / len(idx)
```

```python
def preprocess_features(x, d):
    """Функция для подготовки датасета к обучению.
    Удаляет ненужные признаки, масштабирует данные."""
    
    columns_to_drop = ['6', '21', '25', '70', '33', '44', '59', '65']
    x.drop(columns_to_drop, axis=1, inplace=True)

    scaler = load('scaler.pkl')
    x = scaler.transform(x)
    
    return x  
```

```python
def create_index(x, d, n_cells, nprobe, ind, M = None, ef_construction = None, ef_search = None):
    """Функция для создания и обучения индекса faiss.
    Принимает на вход информацию для обучения индекса, возвращает обученный индекс."""
    
    if ind == 'IndexFlatL2':
        quantizer = faiss.IndexFlatL2(d)
        index = faiss.IndexIVFFlat(quantizer, d, n_cells)
        index.train(x)
    elif ind == 'IndexFlatIP':
        quantizer = faiss.IndexFlatIP(d)
        index = faiss.IndexIVFFlat(quantizer, d, n_cells)
        index.train(x)
    elif ind == 'IndexHNSWFlat':
        index = faiss.IndexHNSWFlat(d, M)
        index.hnsw.efConstruction = ef_construction
        index.hnsw.efSearch = ef_search
    elif ind == 'IP+IndexIVFPQ':
        quantizer = faiss.IndexFlatIP(d)
        index = faiss.IndexIVFPQ(quantizer, d, n_cells, 64, 8)
        index.train(x)
    elif ind == 'IndexLSH':
        index = faiss.IndexLSH(d, 4*d)
        index.train(x)
        

    print('index is_trained')
    faiss.normalize_L2(np.ascontiguousarray(x.values).astype('float32'))
    index.add(x)
    print(f'index.ntotal = {index.ntotal}')
    index.nprobe = nprobe
    
    return index
```

#### Генерируем предсказания Faiss как кандидатов для второго этапа

Для обучения бустинга генерируем новые признаки:
 - расстояние l2
 - разность между векторами
 - добавляем таргет: 1 если поле candidate соответствует targets для данного товара, 0 - если не соответствует

```python
# Загрузим результаты работы модели faiss
vecs_ip = load('vecs_ip_20.pkl')
idx_ip = load('idx_ip_20.pkl')
```

```python
def generate_candidates(x, y, k, vecs, idx, indices):
    """Функция для генерации датасета для обучения на втором этапе.
    Принимает на вход датасет с признаками, соответствующие ему таргеты, k = количество соседей,
    vecs, idx - результат работы faiss, indices - индексы тренировочной выборки"""

    columns = x.columns
    x = x.iloc[np.repeat(np.arange(len(x)), k)]
    
    x['l2'] = vecs[indices].flatten()
    x['candidate'] = idx[indices].flatten()
    
    tqdm.pandas()
    x['candidate_name'] = x['candidate'].progress_apply(lambda x: base.iloc[x].name)
    x[columns] = x.progress_apply(lambda row: row[columns] - base.iloc[row['candidate']], axis = 1)

    y = y[np.repeat(np.arange(len(y)), k)]
    y = (y == x['candidate_name']).astype('int')
    return x, y
```

```python
# Разделим данные на тренировочную и тестовую выборки
indices = [i for i in range(0, len(train))]

X_train, X_test, y_train, y_test, indices_train, indices_test = train_test_split(train, targets, indices, test_size=0.2, 
                                                                                 random_state = RANDOM_STATE)
```

### 2️⃣ 2 этап: обучаем бустинг (`CatBoost`) 

- Ставим задачу как бинарную классификацию
- Выбираем `CatBoostClassifier`

```python
# Подбор гиперпараметров
location = "cachedir"
memory = Memory(location=location, verbose=0)

pipeline = make_pipeline(
        'passthrough', memory=memory
        )

param_grid = [
{   'passthrough': [CatBoostClassifier(random_state=RANDOM_STATE,
                                      class_weights = [1, 20])],  
    'passthrough__thread_count' : [-1],
    'passthrough__verbose' : [500],
    'passthrough__early_stopping_rounds' : [100],
    'passthrough__eval_metric' : ['Precision'],
    'passthrough__learning_rate' : stats.uniform(0.0001, 0.3),
    'passthrough__l2_leaf_reg' : [2, 7],
    'passthrough__depth' : stats.binom(n=10, p=0.2)
}]

search = RandomizedSearchCV(pipeline, param_grid, error_score = 'raise', scoring = 'average_precision', cv = 3)

search.fit(X_train.drop(['candidate', 'candidate_name'], axis = 1), y_train)

memory.clear(warn=False)
rmtree(location)
```

```python
# Лучший оценщик 
model = search.best_estimator_[0]
model
```

```python
# Лучшие параметры модели из перебора
search.best_params_
```

```python
# Сохраним модель
model.save_model('catboost_model.cbm', format="cbm")  
```

Рассчитаем метрику accuracy@5 для X_test

```python
def metric_calculation(model, X, targets):
    """Функция для расчета метрики на основании предсказаний модели
    Принимает на вход обученную модель, данные для предсказаний и наиболее подходящие товары - targets"""
    
    X['preds'] = model.predict_proba(X.drop(['candidate', 'candidate_name'], axis = 1))[:, 1]
    X = X.reset_index()
    res = X.groupby('Id')['candidate'].apply(lambda x: x.head(5).tolist())
    targets = targets.reset_index()
    targets = targets.sort_values(['Id'])
    return (f'acc = {calc_acc(res, targets["Target"], base_index)}')
```

```python
# Рассчитаем метрику accuracy@5 для X_test
metric_calculation(new_clf, X_test, targets[indices_test])
```

**Для обучения бустинга использовались признаки:**
- расстояние l2 между векторами
- покоординатная разность между векторами
В качестве метрики использовался Precision. Балансировка дисбаланса классов осуществлялась за счет взвешивания классов. 

**Самые важные признаки по мнению модели:**
1. Расстояние l2
2. Признаки 45, 5, 52, 8 - это поокординатные разности между соответствующими координатами векторов.

**Что можно улучшить:**

- объединить предсказания нескольких индексов
- обучить catboost с другой метрикой, например, ROC-AUC
